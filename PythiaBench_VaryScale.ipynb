```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Files are raw output from Pythia8245 using fastjet3.3.4.
The naming format sets the pion mass as the first value and mediator as the second in GeV. Looking at keeping the hadronization scale and therefore quark mass constant while varying the spectrum and mediator.
* Model 1: $m_{\pi}=1$ GeV, $m_{X}=750$ GeV, $\Lambda \approx m_{Q} = 10$ GeV
* Model 2: $m_{\pi}=5$ GeV, $m_{X}=750$ GeV, $\Lambda \approx m_{Q} = 10$ GeV
* Model 3: $m_{\pi}=5$ GeV, $m_{X}=1500$ GeV, $\Lambda \approx m_{Q} = 10$ GeV
* Model 4: $m_{\pi}=8$ GeV, $m_{X}=750$ GeV, $\Lambda \approx m_{Q} = 10$ GeV

```python
data_path = "/Users/dylanlinthorne/pythia8245/examples/"
data1 = np.loadtxt(data_path + "MATEngB4_750.dat")
data2 = np.loadtxt(data_path + "MATEngB5_750.dat")
data3 = np.loadtxt(data_path + "MATEngB5_1500.dat")

data1_all = np.loadtxt(data_path + "MATEngB4_750_all.dat")
data2_all = np.loadtxt(data_path + "MATEngB5_750_all.dat")
data3_all = np.loadtxt(data_path + "MATEngB5_1500_all.dat")

dataPt1 = np.loadtxt(data_path + "MATPt4_750.dat-0.dat")
dataPt2 = np.loadtxt(data_path + "MATPt5_750.dat-0.dat")
dataPt3 = np.loadtxt(data_path + "MATPt5_1500.dat-0.dat")
```

Dark Pion boosts; Energy of decay products, at both hadronization and decay scales; pseudorapidty profiles; transverse momentum of leading Jet (dark or sm).

```python
fig, ax = plt.subplots(2,2,sharey=False,figsize=(13,10))

ax[0,0].hist(data1[:,0], 40, (0, 70), alpha=0.75,histtype='step',lw=1.5, label='$m_{\pi}=2$ GeV, $m_{X}=750$ GeV', density = 1);
ax[0,0].hist(data2[:,0], 40, (0, 70), alpha=0.75,histtype='step',lw=1.5, label='$m_{\pi}=5$ GeV, $m_{X}=750$ GeV', density = 1);
ax[0,0].hist(data3[:,0], 40, (0, 70), alpha=0.75,histtype='step',lw=1.5, label='$m_{\pi}=5$ GeV, $m_{X}=1500$ GeV', density = 1);
ax[0,0].legend(loc='upper right');
ax[0,0].set_xlabel('$|p|/m$');

ax[0,1].hist(data1[:,1], 40, (0, 70), alpha=0.75,histtype='step',lw=1.5, label='$m_{\pi}=2$ GeV, $m_{X}=750$ GeV', density = 1);
ax[0,1].hist(data2[:,1], 40, (0, 70), alpha=0.75,histtype='step',lw=1.5, label='$m_{\pi}=5$ GeV, $m_{X}=750$ GeV', density = 1);
ax[0,1].hist(data3[:,1], 40, (0, 70), alpha=0.75,histtype='step',lw=1.5, label='$m_{\pi}=5$ GeV, $m_{X}=1500$ GeV', density = 1);

ax[0,1].hist(data1[:,3], 40, (0, 70), alpha=0.4,histtype='step',lw=1.5,color= 'tab:blue', density = 1);
ax[0,1].hist(data2[:,3], 40, (0, 70), alpha=0.4,histtype='step',lw=1.5,color= 'tab:orange', density = 1);
ax[0,1].hist(data3[:,3], 40, (0, 70), alpha=0.4,histtype='step',lw=1.5,color= 'tab:green', density = 1);

ax[0,1].legend(loc='upper right');
ax[0,1].set_xlabel('Energy [GeV]');


ax[1,0].hist(data1[:,2], 40, (0, 4), alpha=0.75,histtype='step',lw=1.5, label='$m_{\pi}=2$ GeV, $m_{X}=750$ GeV', density = 1);
ax[1,0].hist(data2[:,2], 40, (0, 4), alpha=0.75,histtype='step',lw=1.5, label='$m_{\pi}=5$ GeV, $m_{X}=750$ GeV', density = 1);
ax[1,0].hist(data3[:,2], 40, (0, 4), alpha=0.75,histtype='step',lw=1.5, label='$m_{\pi}=5$ GeV, $m_{X}=1500$ GeV', density = 1);
ax[1,0].legend(loc='upper right');
ax[1,0].set_xlabel('$\eta$');

ax[1,1].hist(dataPt1[:,0], dataPt1[:,2], weights = dataPt1[:,1],  alpha=0.75, histtype='step',lw=1.5, label='$m_{\pi}=2$ GeV, $m_{X}=750$ GeV', density = 1);
ax[1,1].hist(dataPt2[:,0], dataPt2[:,2], weights = dataPt2[:,1],  alpha=0.75, histtype='step',lw=1.5, label='$m_{\pi}=5$ GeV, $m_{X}=750$ GeV', density = 1);
ax[1,1].hist(dataPt3[:,0], dataPt3[:,2], weights = dataPt3[:,1],  alpha=0.75, histtype='step',lw=1.5, label='$m_{\pi}=5$ GeV, $m_{X}=1500$ GeV', density = 1);
ax[1,1].axvline(x=200,linewidth=1.5, alpha=0.6,ls = '--', color='black');
ax[1,1].annotate('$P_{T}$ cut', xy = (10,.0001), xycoords='data',fontsize=14)
ax[1,1].legend(loc='upper right');
ax[1,1].set_xlabel('Leading Jet $p_{T}$');
```
