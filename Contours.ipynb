```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
import scipy as sp

# tranforming json file into a pandas dataframe library
from pandas.io.json import json_normalize

import matplotlib.pylab as plt
plt.rcParams["figure.figsize"] = (8,5)
```

```python
al3x_df = pd.read_csv("EJData/codexb_grid.csv",header=None)
```

```python
columns = ["Pi Mass "+str(x) for x in range(0,61)]
al3x_df.columns = columns
al3x_df.rename(columns={"Pi Mass 0": "Log(tau)"}, inplace=True)

al3x_df
```

```python
x_cross = 3*6.2;
Lumii_full = 3000;
Lumii_semi = 300;
Lumii_small = 250;

al3x_cross = al3x_df.apply(lambda row: 3/row*Lumii_full if row.name != 'Log(tau)' else row, axis=0)
al3x_num = al3x_df.apply(lambda row:row*Lumii_semi*x_cross if row.name != 'Log(tau)' else row, axis=0)
```

```python
x = np.linspace(1,60,60)

#al3x_plot = al3x_cross.drop(columns=['Log(tau)'])
al3x_plot = al3x_cross.set_index('Log(tau)')
#al3x_plot = al3x_plot.drop(7.00)
y=al3x_plot.index.values

#al3x_plot
#y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)

#mylevels=[1.e1, 1.e2, 1.e3, 1.e4, 1.e5, 1.e6, 1.e7, 1.e9 ];

#contours = plt.contour(Y,X, al3x_plot.values, levels= mylevels, cmap='RdGy');
contours = plt.contour(X,Y, al3x_plot.apply(np.log).values,30, antialiased=True);
plt.contourf(X,Y, al3x_plot.apply(np.log).values,5, antialiased=True);


plt.clabel(contours, inline=True, fontsize=8)
plt.ylabel("Log(t)", fontsize =12)
plt.xlabel("$M_{\pi}$ [GeV] ", fontsize = 12)

#plt.ylim(17,20)
y.shape
#al3x_plot.values.shape
```

```python
fig, ax = plt.subplots()
t = ax.axvspan(-2, -1, ymin=0, ymax=1, alpha=.6, color='Black')
ax.axhspan(-1, 1, xmin=0, xmax=1, alpha=.8, color='Grey')

x = np.linspace(1,60,60)
x = sp.ndimage.zoom(x, 1)
#al3x_plot = al3x_cross.drop(columns=['Log(tau)'])
al3x_plot_n = al3x_num.set_index('Log(tau)')
#al3x_plot = al3x_plot.drop(7.00)
y=sp.ndimage.zoom(al3x_plot_n.index.values,1)

#al3x_plot
#y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)

mylevels=[3,30,1000, 5000]

data = sp.ndimage.zoom(al3x_plot_n.values, 1)

#x.shape
#data.shape

contours = plt.contour(Y,X, data, levels= mylevels, antialiased=True);
cs = plt.contourf(Y,X, data, levels= mylevels,  alpha=.5, antialiased=True);

#plt.clabel(contours, inline=True, fontsize=8, colors= 'white')
plt.xlabel("Log($t/1$m)", fontsize =12)
plt.ylabel("$M_{\pi}$ [GeV] ", fontsize = 12)
plt.title("AL3X Contour Test ", fontsize = 14)
plt.grid()
plt.ylim(1,60)
plt.xlim(-2,10)

proxy = [plt.Rectangle((0,0),1,1,fc = pc.get_facecolor()[0]) 
    for pc in cs.collections]
proxy.append(plt.Rectangle((0,0),1,1,fc = 'Grey'))
plt.legend(proxy, ["$N > 3$", "$N > 300$", "$N > 1000$", "$M_{X} < 1$ TeV "])
plt.show()

#y.shape
#al3x_plot.values.shape
fig.savefig('test.pdf')
```
