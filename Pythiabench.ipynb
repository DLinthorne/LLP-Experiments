## Pythia8 Emerging Jet Benchmarks (Constant Scale)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Files are raw output from Pythia8245 using fastjet3.3.4.
The naming format sets the pion mass as the first value and mediator as the second in GeV. Looking at keeping the hadronization scale and therefore quark mass constant while varying the spectrum and mediator. i.e. $\Lambda \approx m_{Q} = 10$ GeV. Keeping the scalar meson masses below this scale. These events are generated using the scalar bifundamental $X$.
* Model 1: $m_{\pi}=1$ GeV, $m_{X}=750$ GeV
* Model 2: $m_{\pi}=5$ GeV, $m_{X}=750$ GeV
* Model 3: $m_{\pi}=5$ GeV, $m_{X}=1500$ GeV
* Model 4: $m_{\pi}=8$ GeV, $m_{X}=750$ GeV

```python
data_path = "/Users/dylanlinthorne/pythia8245/examples/EJData_constant_scale/"
data1 = np.loadtxt(data_path + "MATEngB1_750.dat")
data2 = np.loadtxt(data_path + "MATEngB5_750.dat")
data3 = np.loadtxt(data_path + "MATEngB5_1500.dat")

data1_all = np.loadtxt(data_path + "MATEngB1_750_all.dat")
data2_all = np.loadtxt(data_path + "MATEngB5_750_all.dat")
data3_all = np.loadtxt(data_path + "MATEngB5_1500_all.dat")

dataPt1 = np.loadtxt(data_path + "MATPt1_750.dat-0.dat")
dataPt2 = np.loadtxt(data_path + "MATPt5_750.dat-0.dat")
dataPt3 = np.loadtxt(data_path + "MATPt5_1500.dat-0.dat")
```

Dark Pion boosts; Energy of decay products, at both hadronization and decay scales; pseudorapidty profiles; transverse momentum of leading Jet (dark or sm).

```python
hadron_PDG = np.loadtxt(data_path + "hadrons.dat")
#hadron_PDG2 = np.loadtxt(data_path + "piontostrange.dat")
hadron_PDG2 = np.loadtxt(data_path + "hadrons3.dat")
hadron_PDG3 = np.loadtxt(data_path + "hadrons3.dat")
final_PDG = np.loadtxt(data_path + "final.dat")
final_PDG2 = np.loadtxt(data_path + "final2.dat")
final_PDG3 = np.loadtxt(data_path + "final3.dat")
final_NodetectorPDG = np.loadtxt(data_path + "final_nodetector.dat")
final_NodetectorPDG2 = np.loadtxt(data_path + "final_nodetector2.dat")
final_NodetectorPDG3 = np.loadtxt(data_path + "final_nodetector3.dat")

hadron_counts = np.bincount(hadron_PDG.astype(int))
hadron_counts2 = np.bincount(hadron_PDG2.astype(int))
hadron_counts3 = np.bincount(hadron_PDG3.astype(int))
final_counts = np.bincount(final_PDG.astype(int))
final_counts2 = np.bincount(final_PDG2.astype(int))
final_counts3 = np.bincount(final_PDG3.astype(int))
nodetector_counts = np.bincount(final_NodetectorPDG.astype(int))
nodetector_counts2 = np.bincount(final_NodetectorPDG2.astype(int))
nodetector_counts3 = np.bincount(final_NodetectorPDG3.astype(int))

index = np.nonzero(hadron_counts)[0]
index2 = np.nonzero(hadron_counts2)[0]
index3 = np.nonzero(hadron_counts3)[0]
indexf = np.nonzero(final_counts)[0]
indexf2 = np.nonzero(final_counts2)[0]
indexf3 = np.nonzero(final_counts3)[0]
indexd = np.nonzero(nodetector_counts)[0]
indexd2 = np.nonzero(nodetector_counts2)[0]
indexd3 = np.nonzero(nodetector_counts3)[0]

Labels = ['$\pi^0$','$\pi^+$', '$\eta$',r'$\rho^{0}$', r'$\rho^{+}$', '$\omega$','K$^0$','K*$^0$','K$^+$','K*$^+$',
     '$\eta\prime$','$\phi$','$\Delta^-$','$n$','$\Delta^0$','$p$','$\Delta^+$', '$\Lambda$']

Labelfs = ['$e^-$',r'$\nu_e$','$\mu^-$',r'$\nu_{\mu}$','$\gamma$','K$_L$','$\pi^+$','K$_S$','K$^+$',
           '$n$','$p$','$\Delta^+$', '$\Lambda$']

insert = [11, 12, 13,  14,  22, 130, 211, 310]
insertd2 = [11,   12,   13,   14,   22,  130,  211,  310,  321, 2112, 2212]
insertd = [11,  12,  13 , 14,  22, 130]

label_pos = [i for i, _ in enumerate(index)]
label_pos2 = [i for i, _ in enumerate(index2)]
label_pos3 = [i for i, _ in enumerate(index3)]
label_posf = [i for i, _ in enumerate(insert)]
label_posf2 = [i for i, _ in enumerate(indexf2)]
label_posf3 = [i for i, _ in enumerate(indexf3)]
label_posd = [i for i, _ in enumerate(insertd)]
label_posd2 = [i for i, _ in enumerate(insertd2)]
label_posd3 = [i for i, _ in enumerate(insertd2)]

fig4, ax4 = plt.subplots(3,1,figsize=(13.5,14))
ax4[0].bar([x+.3 for x in label_pos3],hadron_counts3[index3],0.3,alpha=0.8,label='$m_{\pi}=8$ GeV, $m_{Q} = \Lambda = 10$ GeV')
ax4[0].bar([x for x in label_pos2],hadron_counts2[index2],0.3,alpha=0.8,label='$m_{\pi}=5$ GeV, $m_{Q} = \Lambda = 10$ GeV')
ax4[0].bar([x-.3 for x in label_pos], hadron_counts[index],0.3,alpha=0.8,label='$m_{\pi}=1$ GeV, $m_{Q} = \Lambda = 10$ GeV')
ax4[0].legend(loc='upper right');
ax4[0].set_title('Primary Decay Products of Dark Pion at Hadronization Scale');
ax4[0].set_xticks(label_pos3);
ax4[0].set_xticklabels(Labels);
ax4[0].set_xlim(-1, 18);

ax4[1].bar([x+.3 for x in label_posf3],final_counts3[indexf3],0.3,alpha=0.8,label='$m_{\pi}=8$ GeV, $m_{Q} = \Lambda = 10$ GeV')
ax4[1].bar([x for x in label_posf2],final_counts2[indexf2],0.3,alpha=0.8,label='$m_{\pi}=5$ GeV, $m_{Q} = \Lambda = 10$ GeV')
ax4[1].bar([x-.3 for x in label_posf], final_counts[insert],0.3,alpha=0.8,label='$m_{\pi}=1$ GeV, $m_{Q} = \Lambda = 10$ GeV')
ax4[1].annotate('$\pi^{\pm}, K^{\pm}, \mu^{\pm}$ decays allowed', xy = (-.5,105000), xycoords='data',fontsize=14, color='black');
ax4[1].legend(loc='upper right');
ax4[1].set_title('Dark Pion Final Decay Products within ATLAS Cylinder');
ax4[1].set_xticks(label_pos3);
ax4[1].set_xticklabels(Labelfs);
ax4[1].set_xlim(-1, 13);

ax4[2].bar([x+.3 for x in label_posd3],nodetector_counts3[insertd2],0.3,alpha=0.8,label='$m_{\pi}=8$ GeV, $m_{Q} = \Lambda = 10$ GeV')
ax4[2].bar([x for x in label_posd2],nodetector_counts2[insertd2],0.3,alpha=0.8,label='$m_{\pi}=5$ GeV, $m_{Q} = \Lambda = 10$ GeV')
ax4[2].bar([x-.3 for x in label_posd],nodetector_counts[insertd],0.3,alpha=0.8,label='$m_{\pi}=1$ GeV, $m_{Q} = \Lambda = 10$ GeV')
ax4[2].annotate('$\pi^{\pm}, K^{\pm}, \mu^{\pm}$ decays allowed', xy = (-.5,190000), xycoords='data',fontsize=14, color='black');
ax4[2].legend(loc='upper right');
ax4[2].set_title('Dark Pion Final Decay Products with No Detector Cut');
ax4[2].set_xticks(label_pos3);
ax4[2].set_xticklabels(Labelfs);
ax4[2].set_xlim(-1, 13);
```
